# Series
1차원 데이터 (정수, 실수, 문자열 등)

## Series 객체 생성
예) 1월부터 4월까지 평균 온도 데이터 (-20, -10, 10, 20)

```python
temp = pd.Series([-20, -10, 10, 20])
temp
```

```python
temp[0] # 1월 온도
```

# Series 객체 생성 (Index 지정)

```python
temp = pd.Series([-20, -10, 10, 20], index=['Jan', 'Feb', 'Mar', 'Apr'])
temp
```

```python
temp['Jan'] # index Jan 에 대핟앟는 데이터 출력
```

```python
# temp['Jun'] # 존재하지 않은 키는 에러
```

# 2. DataFrame
2차원 데이터 (Series들의 모음)

## Data 준비
사전 (dict) 자료구조를 통해 생성

예) 슬럼덩크 주요 인물 8명에 대한 데이터

```python
data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
data
```

```python
type(data)
```

```python
data['이름']
```

## DataFrame 객체 생성

```python
import pandas as pd

df = pd.DataFrame(data)
df
```

```python
df['이름']
```

```python
df[['이름', '키']]
```

## DataFrame 객체 생성 (Index 지정)

```python
df = pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번', '6번', '7번', '8번']) # 개수 안맞으면 에러
df
```

## DataFrame 객체 생성 (Column 지정)
data 중에서 원하는 column만 선택하거나, 순서 변경 가능

```python
df = pd.DataFrame(data, columns=['이름', '학교', '키']) # 원하는 컬럼 선택해서 가져 올 수 있음
df
```

```python
df = pd.DataFrame(data, columns=['키', '학교', '이름']) # 순서변경도 ok
df
```

# 4. 파일 저장 및 열기
DataFrame 객체를 excel, csv, txt 등 형태의 파일로 저장 및 열기

```python
import pandas as pd

data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}

df = pd.DataFrame(data, index=['1번', '2번', '3번', '4번', '5번', '6번', '7번', '8번'])
df.index.name = '지원번호'
df
```

# 9. 데이터 선택 (조건)
조건에 해당하는 데이터 선택

```python
import pandas as pd

df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['키'] >= 185 # 학생들의 키가 185이상인지 여부를 True / False
```

```python
df[df['키'] <= 185]
```

```python
df[~(df['키'] <= 185)] # 거꾸로~
```

```python
df.loc[df['키'] >= 185, '수학']
```

## 다양한 조건

### & 그리고

```python
df.loc[(df['키'] >= 185) & (df['학교'] == '북산고')] # 괄호 감싸줘야댐
```

### | 또는

```python
df.loc[(df['키'] < 170) | (df['키'] > 200)]
```

### str 함수

```python
filt = df['이름'].str.startswith('송') # 송으로 시작하는 이름
df[filt]
```

```python
filt = df['이름'].str.contains('태') # 이름에 '태'가 들어가는 사람
df[filt]
```

```python
df[~filt] # 이름에 태가 들어가는사람 제외
```

```python
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs) # SW특기가 Python이거나 Java인 사람
df[filt]
```

```python
langs = ['python', 'java']
filt = df['SW특기'].str.lower().isin(langs) # 소문자로 바꿔서 비교
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java')
print(filt) # True False가 아닌 Nan으로 데이터가 남기때문
df[filt] # Nan이 있는 경우 contains함수는 error
```

```python
filt = df['SW특기'].str.contains('Java', na=True) # Nan데이터에 대해서 True로 설정
df[filt]
```

```python
filt = df['SW특기'].str.contains('Java', na=False) # Nan데이터에 대해서 True로 설정
df[filt]
```

# 10. 결측치
비어 있는 데이터

```python
import pandas as pd

df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

## 데이터 채우기 fillna

```python
df.fillna('') # Nan 데이터를 빈칸으로 채움
```

```python
df.fillna('없음')
```

```python
import numpy as np

df['학교'] = np.nan # 학교 데이터 전체를 Nan으로 채움
df
```

```python
df.fillna('모름')
```

```python
df.fillna('모름', inplace=True) # 업데이트
df
```

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df['SW특기'].fillna('확인 중', inplace=True) # SW특기 데이터 중 Nan에 데이터를 채움
df
```

## 데이터 제외하기 dropna

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df.dropna() # Nan를 포함하는 데이터는 삭제
```

```python
# axis: index or columns (기준)
# how : any or all (하나라도 Nan, 전체다 Nan)
df.dropna(axis='index', how='any') 
```

```python
df.dropna(axis='columns', how='any') 
```

```python
df['학교'] = np.nan
df.dropna(axis='index', how='all')
```

```python
df.dropna(axis='columns', how='all')
```
