# Hypothesis using matrix


$$
H(X) = XW
$$

dot product(=scalar product, 내적)을 통해 $H(x_1, x_2, x_3)=w_1x_1+w_2x_2+w_3x_3+b$연산을 보다 쉽게 컴퓨터로 계산한다. 매개 변수개수가 3개 일 경우는 일일히 식을 써주겠지만 10개, 20개, 100개가 넘어가게 된다면 행렬을 사용한 내적을 통한 연산이 보다 유리하다는 것이다.
